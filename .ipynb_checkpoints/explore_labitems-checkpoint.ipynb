## Q: what are most commonly logged labs? 
Lets pick the top 100 labs for now that are most commonly logged.

```python
#get lab events

query = query_schema + """

WITH top_lab_events AS (

    SELECT itemid, COUNT(*)

    FROM
         labevents

    GROUP BY
        itemid

    ORDER BY 
        count DESC

    LIMIT 100
)
SELECT b.itemid
    ,label
    ,fluid
    ,category
    ,b.count AS lab_count

FROM 
    d_labitems a
    
INNER JOIN top_lab_events b

ON a.itemid = b.itemid

ORDER BY lab_count DESC
"""
df = pd.read_sql_query(query,con)

df

```
